시간 영역 분석부는 측정된 근전도 신호의 면적을 바탕으로 각 사이클의 적분 근전도 값을 수행 시간으로 나눠 평균 적분 근전도 값을 구한다. 그리고 그 평균 적분 근전도 값의 증가율을 계산하는 기능을 수행한다.  
$\rightarrow$ **증가량 계산모듈**은 **근전도 계산 모듈**로 계산된 평균 적분근전도 값의 증가율을 계산한다.

주파수 영역 분석부는 근육 피로 정보 확인을 위해 측정된 근전도 신호를 활용해 주파수 스펙트럼을 생성하고, 중심 주파수 값을 계산한다. 그리고 중심 주파수 값의 감소율을 계산하는 기능을 수행한다.  
$\rightarrow$ 구체적으로 **변환모듈**에서 고속 푸리에 변환하여 주파수 스펙트럼을 생성하고, **중심 주파수값 계산 모듈**에서 중심 주파수를 계산해 주파수 변화 정도를 관찰한다. 근 피로도 축적에 따라 저주파수 대역에서 진폭이 증가하는 정도를 보는 것이다. **감소율 계산모듈**은 그 이후 계산된 중심 주파수 값의 감소율을 계산한다.  

평가부는 위의 두 영역부에서 계산된 평균 적분근전도 값의 증가율, 중심 주파수 감소율을 바탕으로 근 피로도와 근력을 평가하고 그 정보를 이용해 평가 지표를 도출한다.  
$\rightarrow$ 평가 시, 근 피로도가 축적되면 무게에 관계없이 평균 적분 근전도 값이 일정한 값으로 수렴하며, 중심 주파수 값이 일정하게 초깃값의 40%가 감소한다. 여기서 초기 중심 주파수 값의 약 60%까지 감소하는 시점이 근 피로도 축적으로 인해 파워가 감소하기 시작하는 시점으로 판단할 수 있다.  
$\rightarrow$ 무게에 따라 평균 적분 근전도의 초깃값이 증가하고, 증가율이 감소하므로 평균 적분근전도 초깃값이 증가하며 증가율이 감소할 경우 반복 무게로 인한 힘이 감소하는 근피로도가 축적된 시점으로 평가할 수도 있다.

근육 피로는 근육 파워의 감소를 야기한다고 가정할 때, 주어진 반복 동작에 대해 제시된 속도를 더이상 따라가지 못하는 시점을 근 피로도 축적의 판단 기준으로 설정한다.  

여기 나오는 데이터는 피험자가 MVC(Maximum Voluntary Contraction)의 20%, 35%, 50%, 75% 조건으로 3$sec$/cycle로 덤벨 컬 동작을 수행한 이두근 근전도 신호이다.

각 사이클에 대해 시간 영역에서 적분근전도 값을 구하고, 주파수 영역에서 푸리에 변환하여 중심 주파수 값을 계산했다. 동작 수행 시간의 영향을 무시하기 위해 각 동작 수행 시간으로 나누어 정규화된 평균 적분근전도 값을 이용했다.  

아래 그래프를 보면 반복 부하 하중(20% MVC)로 인한 근 피로도가 축적됨에 따른 변화를 보여주고 있다. 덤벨 무게가 증가할수록 근전도 신호의 진폭(Amplitude)이 증가함을 확인할 수 있다. 이런 증가는 더큰 무게에 대해 더 많은 근육이 사용되면서 근섬유 활동 전위가 증가했기 때문이다. 또 피로가 축적됨에 따라 신호의 진폭이 점점 커지는데, 피로해지면서 근섬유 수축 능력이 감소해 운동 단위의 동시화(Synchronization)와 점증(recruitment)이 발생했기 때문이다.
