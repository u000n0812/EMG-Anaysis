운동 부하는 무게에 의해 사용자에게 가해지는 중량이므로, 복수 근육을 사용하며 운동할 시에는 동원 근력과 일치하지 않을 수 있다. 따라서 1RM은 운동 부하를 기준으로 종목별로 선정되는 **운동 1RM**과 동원 근력을 기준으로 근육 부위별로 선정되는 **근육 1RM**으로 나눈다.  



적분근전도(IEMG)와 RMS 데이터를 얻고 이 정보왕 동원 근력 사이의 상관성을 정립할 수 있었다.  

단위 시간동안 수행한 운동량인 근파워(Muscle Power)는 동원 근력과 근수축 속도에 의해 얻어진다.  

**근육 1RM**은 1회 수축해 최대로 동원할 수 있는 동원 근력에 해당하고, 1회 운동에서 최대로 발휘할 수 있는 운동 부하를 **운동 1RM**으로 산정한다.

**운동 1RM**은 **간접 측정법**으로 측정한다. 약간의 부하가 가해지는 운동 기구를 이용해 최대한 반복 운동해서 이때의 부하와 횟수로 1RM을 산정한다.  

$1RM = W(1+0.025R)$  
W : 무게  
R : 최대 반복 횟수

하지만 W를 변경해가면서 얻는 최대 반복 횟수에 근거해 통계적 기법으로 1RM을 산정하는 방식을 많이 사용하므로, 위의 수학식보다 데이터에 근거한 통계 기법을 적용할 수도 있다.  

$\rightarrow$ 두 방법 중 하나를 사용한 것 같다

- 최대 근력 강화 : 최대 근력을 강화하는 근력 운동
- 근 비대 : 근육량 증가
- 근 파워 : 근수축 속도를 반영한 운동 능력. 순발력 향상
- 근 지구력 : 오랜 시간 일정한 근력을 지속적으로 발휘하는 능력

위 표는 데이터가 수집됨에 따라 계속 수정될 것이다. 그리고 겹치는 부분이 있으면 묶어서 하나의 카테고리로 분류할 수도 있다.  

근력 운동 시 운동 리듬을 빠르게 할수록 근 파워를 효과적으로 강화할 수 있는데, 이에 따라 운동 속도 뿐 아니라 근전도 파형 분석을 통해 **운동 리듬(수축/이완 속도, 세트 내 반복 수행하는 운동 주기, 1회 운동할 때 단계별 동작 소요시간 등)** 을 얻는다.

여기서는 동원 근력이 근력 운동 효과의 특정 범위에 도달한 횟수 또는 머문 시간을 누적 점수로 계산하여, 각 근육에 대해 근력 운동 효과별 누적 효과 점수를 계산한다. 이를 상홍 비교해 크기 차이에 따라 불평형 점수를 계산한다.  

이때 운동 효과별 불평형 점수는 동일 근력 운동 효과의 누적 효과 점수 최대치 대비 차이의 비율이다. 최종 점수 계산 시에는 누적 점수 크기에 따른 가중치를 두어 백분율 값으로 얻는다.

UI는 근육별 동원 근력, 근파워 및 근 피로도, 근육별 4가지 누적 효과 점수 및 불평형 점수를 보여준다. 실시간으로 효과를 알 수 있게 하며, 좌우 밸런스를 맞출 수 있게 한다.  

동원 근력을 1RM 대비 비율로 할 수도 있으나, 직관적으로 이해하기 쉽게 $kg/s$ 또는 $Nm/s$*(뉴턴)* 단위로 표시한다.  

효과 표시 텍스트는 최대 근력 강화 효과를 얻을 수 있는 동원 근력 범위에 **'S'**, 근 비대 효과에는 **'H'**, 근 지구력 효과에는 **'E'** 로 표기했다. **'E'** 대신 **'P'** 를 써도 무방하다.

막대 그래프는 동원 근력 축 레이블은 1RM 대비 동원 근력 비율로 표기했고(중량 단위로 해도 상관없다), 근력 축에는 근력 크기에 따라 얻는 근력 운동 효과가 표기되어 있다.  

눈금으로 구분되는 구획 구간에는 누적 점수 텍스트가 표시되어 있으며, 각 구간에 대응되는 누적 효과 점수이다. 누적 효과 점수를 누적 점수 텍스트로, 불평형 점수를 불평형 점수 텍스트로 표출한다.  

누적 효과 점수는 사용자가 횟수/시간으로 선택한 형식으로 나타낸다.
