여기서는 어깨와 손목 사이의 상지 근육에 대한 근전도를 측정했다. 재활 운동을 하고 있는 사용자의 상지 근육에 부착된 센서로부터 신호를 얻었다. 근전도 신호는 사용자의 운동에 따라 바뀌는 주파수와 진폭을 갖는다.  

근전도 신호의 **진폭을 이용해 상지 근육의 활성도를 연산**할 수도 있다. 예를 들어 근전도 신호의 진폭에 대한 **진폭확률분포함수(Amplitude Probability Distribution Function ; APDF)**  를 생성할 수 있고, 그 다음 근전도 신호의 진폭에 대한 **APDF를 복수의 백분위수로 나누어 활성도를 연산**할 수 있다.

아래 그림을 보자.  

X축은 근전도 신호의 자발적 수축(Reference Voluntary Contraction ; RVC)의 백분율을 나타내고, Y축은 rvc 백분율에 대한 진폭 확률(Amplitude probability)을 나타낸다. 예를 들어 10번째 백분위수는 근전도 신호의 진폭 크기를 정렬한 다음 진폭 확률이 10번째 순위에 해당하는 RVC의 백분율을 나타낸다.  

따라서 RVC 백분율을 기준으로 활성도를 연산하는게 아니라, 진폭 확률에 따른 백분위수를 이용해 그전도 신호의 진폭이 특정 백분위수에 해당하는 RVC의 백분율을 연삼함으로써 다양한 근육 활성화 분포를 확인할 수 있다.

재활 운동의 측면에서, 주동근과 길항근이 동시에 수축하는지 여부에 따라 활성도를 계산할 수도 있다. 동시 수축하는 경우 근육 활성도에 반영하는 식이다. 이는 근육의 안정성, 기민성, 정확성 등의 활동 개선 여부를 평가하는 지표로 활용된다.  

위의 아래 그래프는 제1근육과 제2근육 신호의 운동 데이터에 따른 진폭 변화를 나타낸다. 운동 데이터는 재활 훈련에 따라 근육을 자극시키는 데이터를 말한다. 따라서 특정 운동 데이터 영역에서 운동 자극에 근육이 반응하면 근전도 신호에서 진폭이 상승하는 것을 확인할 수 있다. 제1 근육진폭 대비 제2근육진폭 신호가 운동 데이터 2000 내지 9000 사이에서 동시 수축하는 것을 확인할 수 있다. 이 경우, 동시 수축 영역 넓이를 적분해 운동 데이터 수로 나눠 동시수축도에 대한 값을 결정할 수 있다. 

피로도 연산은 **파워 스펙트럼 분석(Power Spectrum Analysis)** 을 통해 **피크 파워(Peak power)**, **중간 주파수(Median frequency)**, **중심 주파수(Mean frequency)**, **전체 파워(Total power)** 를 추출할 수 있다.  

이중 중간 주파수를 주로 활용해 근육의 피로도를 연산하며, 근전도 신호의 주파수 중 피로도에 따라 미리 설정된 크기의 저주파 성분이 증가하는 지점을 이용해 피로도를 연산한다. 즉 근육이 피곤해질수록 근전도 신호 주파수가 고주파에서 저주파 성분이 증가하고, 중간 주파수가 낮아지는 특성을 이용하는 것이다.  

진폭확률분포함수와 동시수축도를 함께 이용하면 활성화 정도에 따른 재활이 어느 정도 진행되었는지 판단할 수 있고, 주파수 신호를 이용해 피로도를 분석함으로써 재활 정도를 판단할 수 있다.
